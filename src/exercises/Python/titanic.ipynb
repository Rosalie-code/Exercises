```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
df.head()
```

```python
"""Probability of survival depending of the class"""
pb_1= (df[df['pclass']==1]['survived'].sum())/df[df['pclass']==1]['pclass'].count()
print(pb_1)
pb_2= (df[df['pclass']==2]['survived'].sum())/df[df['pclass']==2]['pclass'].count()
print(pb_2)
pb_3= (df[df['pclass']==3]['survived'].sum())/df[df['pclass']==3]['pclass'].count()
print(pb_3)
```

```python
"""Probability of survival depending of the age and sex"""
pb_m= (df[df['who']=="man"]['survived'].sum())/df[df['who']=='man']['who'].count()
print(pb_m)
pb_w= (df[df['who']=='woman']['survived'].sum())/df[df['who']=='woman']['who'].count()
print(pb_w)
pb_m_c= (df[(df['who']=='child') & (df['sex']== 'male')]['survived'].sum())/df[(df['who']=='child')&(df['sex']== 'male')]['who'].count()
print(pb_m_c)
pb_m_w= (df[(df['who']=='child') & (df['sex']== 'female')]['survived'].sum())/df[(df['who']=='child')&(df['sex']== 'female')]['who'].count()
print(pb_m_w)
```

```python
"""P(A/B)"""

pb_A= (df[df['adult_male']==True]['survived'].sum())/df[df['adult_male']==True]['adult_male'].count()
print(pb_A)

pb_1_B= (df[df['pclass']==1]['survived'].sum())/df[df['pclass']==1]['pclass'].count()
print(pb_1)
pb_A_1_B= (df[(df['adult_male']==True) & (df['pclass']==1)]['survived'].sum())/df[(df['adult_male']==True) & (df['pclass']==1)]['pclass'].count()       
print((pb_A_1_B)/(pb_1_B))
```

```python
"""Average Age of Passengers"""

survived = df[df['survived']==1]['age'].mean()
not_survived = df[df['survived']==0]['age'].mean()
print(survived)
print(not_survived)
```

```python
"""Average Fare of Passengers"""

survived = df[df['survived']==1]['fare'].mean()
not_survived = df[df['survived']==0]['fare'].mean()
print(survived)
print(not_survived)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['fare'], alpha=0.5, c='blue')
plt.title('Relation between Age and Fare Paid of Passengers')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = {1: 'blue', 2: 'green', 3: 'orange'}
df['color'] = df['pclass'].map(colors)

plt.figure(figsize=(10, 6))
for pclass in [1, 2, 3]:
    subset = df[df['pclass']== pclass]
    plt.scatter(subset['age'], subset['fare'], alpha=0.5, c=subset['color'], label=f'pclass {pclass}')
plt.title('Relation between Age and Fare Paid of Passengers')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```
